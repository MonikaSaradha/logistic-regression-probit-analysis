# Logistic & Probit Regression Analysis

Heart disease prediction and consumption pattern analysis using classification models.

```
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, roc_curve, auc
```

## Heart Disease Dataset

```
heart = pd.read_csv("heart.csv")
X = heart.drop('output', axis=1)
y = heart['output']
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### Logistic Regression

```
log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train, y_train)
y_pred = log_model.predict(X_test)

print(confusion_matrix(y_test, y_pred))
```

```
y_prob = log_model.predict_proba(X_test)[:,1]
fpr, tpr, _ = roc_curve(y_test, y_prob)
plt.plot(fpr, tpr)
plt.title("ROC - Logistic Regression")
plt.show()
```

### Decision Tree

```
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
y_pred_tree = tree.predict(X_test)

print(confusion_matrix(y_test, y_pred_tree))
```

## NSSO Dataset - Probit Analysis

```
import statsmodels.api as sm

data = pd.read_csv("punjab.csv")

features = ['eggsno_q','fishprawn_q','goatmeat_q','beef_q','pork_q','chicken_q','othrbirds_q']
X = data[features]
y = (data['eggsno_q'] > 0).astype(int)

X = sm.add_constant(X)

model = sm.Probit(y, X).fit()
print(model.summary())
```
